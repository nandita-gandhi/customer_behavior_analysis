### Now to impute the missing values - we can use median - as it is robust towards outliers and also very close to mean value. But this isn't a best practice as the review rating of different categories are there and if we impute a single media value to all missing data - the data will be biased. To overcome this, we will first group the review ratings with categories and then impute the median of that into missing values.

```python
df["Review Rating"] = df.groupby('Category')['Review Rating'].transform(lambda x: x.fillna(x.median()))
```

```python
df.isnull().sum()
```

### Feature Engineering

### 1. create a column age_group to group ages of customers into 4 groups (Numeric into Categorical)

```python
labels = ['Young Adult','Adult','Middle-aged','Senior']
```

### use qcut - it splits the ages into 4 equal size groups based on the data distribution and assigns the labels we defined

```python
df['age_group'] = pd.qcut(df['age'], q = 4, labels = labels)
```

```python
df[['age','age_group']].head(10)
```

### 2. create a column purchase_frequency_days to have number of data instead of categorical to make the frequency feature understandable for analysis (Categorical into Numeric)

```python
frequency_mapping = {
    'Fortnightly': 14,
    'Weekly': 7,
    'Monthly': 30,
    'Quarterly': 90,
    'Bi-Weekly': 14,
    'Annually': 365,
    'Every 3 Months': 90
}
df['purchase_frequency_days'] = df['frequency_of_purchases'].map(frequency_mapping)
```
